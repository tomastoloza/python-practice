Dado el siguiente fragmento de código que implementa el núcleo de un sistema de gestión de cuentas bancarias, crear las excepciones que correspondan para verificar que se cumplen las precondiciones y los invariantes de cada función. Escribir los handlers en las funciones que llaman a las que pueden levantar una excepción para que la manejen y muestren un mensaje de error acorde, citando información de contexto (número de cuenta, saldo, etc.).
Sacar la marca de comentario de @preconditions, para evitar cuentas con saldo menor que cero

```python
import time
from preconditions import preconditions

class Cuenta(object):
    """Clase cuenta, modela una cuenta bancaria básica
       define el comportamiento para acreditar y debitar montos
       Cada cuenta guarda todos los movimientos que se realizan
       @invariant: La suma de todas las operaciones realizadas
       debe ser igual al saldo
       @invariant el saldo de la cuenta no puede ser negativo
    """
#    @preconditions(
#        lambda monto_inicial: isinstance(monto_inicial, float),
#        lambda monto_inicial: monto_inicial >=0,
#        )    

    def __init__(self, numero, monto_inicial=0.0):
        """
        @type numero        : string
        @param numero       : numero de cuenta
        @type monto_inicial : float
        @param monto_inicial: dinero con que se abre una cuenta
        """
        self.__numero__=numero
        self.__monto_inicial__=monto_inicial
        self.__saldo__=0.0
        self.__movimientos__=[]
        self.__realizar_transaccion__('Deposito inicial',\
                                      monto_inicial,\
                                      'credito')
    def get_saldo(self):
        """
        @return: el saldo actual de la cuenta
        """
        return self.__saldo__
    
    def get_monto_inicial(self):
        """
        @return: el monto inicial con el que se abrio la cuenta
        """
        return self.__monto_inicial__
    
    def get_numero(self):
        """
        @return: el nro de cuenta
        """
        return self.__numero__

    def __realizar_transaccion__(self, nombre, monto, tipo):
        """Solo hay dos tipos de transacciones: credito y debito
        @type nombre : string
        @param nombre: nombre descriptivo del movimiento, por ej. retiro ATM
        @type monto  : float
        @param monto : monto involucrado en la transaccion
        @type tipo   : string
        @param tipo  : tipo de transaccion: credito o debito
        @todo        : agregar excepciones tipos invalidos de transaccion
        """
        if tipo == 'credito':
            self.__saldo__=self.__saldo__ + monto
            transaccion=Transaccion(nombre, monto, tipo, time.localtime())
            self.__movimientos__.append(transaccion)
        elif tipo == 'debito':
            self.__saldo__=self.__saldo__-monto
            transaccion=Transaccion(nombre, -monto, tipo, time.localtime())
            self.__movimientos__.append(transaccion)
        else:
            pass
            
    def __rollback__(self):
        """Deshace la ultima transaccion
        
        """
        t=self.__movimientos__.pop()
        monto=abs(t.monto)
        if t.tipo == 'debito':
            self.__saldo__=self.__saldo__ + monto
        elif t.tipo == 'credito':
            self.__saldo__=self.__saldo__ - monto
        else:
            pass    

    def imprimir_movimientos(self):
        """Muestra por pantalla todos los movimientos realizados en
        la cuenta
        """
        print("Movimientos de la cuenta {0:<10s} a la fecha (SEUO):\n".\
              format(self.get_numero()[:10]))
        print('{0:<17s}{1:<10s}{2:>10}'.\
              format('Fecha','Operacion','Monto'))
        for mov in self.__movimientos__:
            print('{0:<17s}{1:<13s}{2:8.2f}'.\
                  format(time.strftime('%d/%m/%Y %H:%M',mov.timestamp),\
                         mov.nombre[:12],\
                         mov.monto))
        print('\n\n{0:<20s}{1:8.2f}'.format("Saldo actual:",\
                                           self.get_saldo()))
    def depositar(self,nombre, monto):
        """Realizar un deposito
        @type nombre   : string
        @param nombre  : nombre descriptivo del movimiento, por ej. retiro ATM
        @type monto    : float
        @param monto   : monto a depositar
        @precondition  : monto>0
        @postcondition : saldo = saldo anterior + monto
        """
        self.__realizar_transaccion__(nombre, monto, 'credito')
        return self.get_saldo()

    def retirar(self,nombre, monto):
        """Realizar una extraccion
        @type nombre  : string
        @param nombre : nombre descriptivo del movimiento, por ej. retiro ATM
        @type monto   : float
        @param monto  : monto a retirar
        @precondition : monto > 0
        @postcondition: saldo = saldo anterior - monto
        """
        self.__realizar_transaccion__(nombre, monto, 'debito')
        return self.get_saldo()

    def transferir_a(self, otraCuenta, monto):
        """Realizar una transferencia a otra cuenta
        @type otraCuenta: Cuenta
        @param otraCuenta: Cuenta destino donde se acreditara el monto
        @type monto   : float
        @param monto  : monto a transferir
        @precondition : monto > 0
        @precondition : otraCuenta existe
        @postcondition: saldo = saldo anterior - monto
        
        """
        self.retirar('transf. a {0}'.format(otraCuenta.get_numero()), monto)
        otraCuenta.depositar('transf. desde {0}'.format(self.get_numero())\
                             , monto)
        return self.get_saldo()



class Transaccion(object):
    """Modela Transaccion: operacion realizada
    """
    def __init__(self,nombre='', monto=0.0, tipo='', fecha=''):
        self.nombre=nombre
        self.monto=monto
        self.tipo=tipo
        self.timestamp=fecha

if __name__ == "__main__":
    cta=Cuenta("1212")
    cta2=Cuenta("2121", 1000)

    cta.depositar("Deposito por ventanilla", 1000.10)
    cta.depositar("Depósito ATM", 1241)
    cta.transferir_a(cta2, 2241.10)
    cta.imprimir_movimientos()
    cta2.imprimir_movimientos()

    ###Empiezan los problemas###
    #Extracción mayor al saldo disponible#
    cta.transferir_a(cta2, 1000)
    
    #Retiro negativo permite incrementar el saldo#
    cta2.retirar("Ret. Ventanilla", -10000)

    cta.imprimir_movimientos()
    cta2.imprimir_movimientos()
    
    cta.__rollback__()
    cta2.__rollback__()
    
    cta.imprimir_movimientos()
    cta2.imprimir_movimientos()
```
